```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dat = pd.read_csv("perpend_eyecenter.csv",sep=',')
```

```python
dat = np.genfromtxt("perpend_eyecenter_np.csv",delimiter=',')
datt = dat.T
```

```python
a = - datt[1] / datt[0]
c = datt[1]
```

```python
org = np.array((a,c))
```

```python
org_c = np.array((a,c,np.arange(len(org[1]))))
```

```python
def draw(a,b):
    plt.figure(figsize=(10,10))
    #plt.ylim([-300,300])


    for i in (np.arange(a, a+b)):
        plt.plot(x,org_c[0][i]*x + org_c[1][i])
    plt.show()
    
def draw_so(a,b):
    x = np.linspace(250,350,100)
    #plt.ylim([-300,300])
    plt.figure(figsize=(10,10))
    for i in (np.arange(a, a+b)):
        plt.plot(x,org_c[0][i]*x + org_c[1][i])
    plt.show()
    
def draw_sop(a,b,c,d):
    #b, d is delta a and c
    x = np.linspace(c,c+d,100)
    plt.figure(figsize=(10,10))
    #plt.ylim([-300,300])
    for i in (np.arange(a, a+b)):
        plt.plot(x,org_c[0][i]*x + org_c[1][i])
    plt.show()

def draw_sopp(a,b,c,d):
    #b, d is delta a and c
    x = np.linspace(c,c+d,100)
    plt.figure(figsize=(10,10))
    #plt.ylim([-300,300])
    for i in (np.arange(a, a+b)):
        plt.plot(x,org_c[0][i]*x + org_c[1][i])
    plt.scatter()
    plt.show()
```

```python
draw(120,60)
```

```python
draw_so(0,6)
```

```python
draw_sop(0,1000,0,640)
```
