```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
k = 20
x = np.linspace(-100,100,200)
va = np.random.randint(0,360, size = k)
b = np.random.randint(-70, 70, size = k) # 와이 절편
a = np.sin(va)/np.cos(va) # 기울기
```

```python
y = 0
for i in range (k):
    y += b[i] * np.abs(x - a[i])
```

```python
A = np.zeros((30,30))
for i in range(len(A)):
    for j in range(len(A)):
        if i<10 and j<10:
            A[i][j]=1
```

```python
x = np.linspace(-1,1,30)
```

```python
xx, yy = np.meshgrid(x,x)
```

```python
plt.scatter(xx, yy, c = A, cmap="hot_r")
plt.colorbar()
plt.show()
```
